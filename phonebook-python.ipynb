```python
#Criando função para adicionar um contato

def adicionarContato():
  nome = input('\n' + 'Digite o nome do contato: ')
  telefone = input('Digite seu número de telefone: ')
  email = input('Digite seu e-mail: ')
  twitter = input('Digite seu Twitter: ')
  instagram = input('Digite seu Instagram: ')
  index = 1
  for n in agenda.keys():
    if n == index:
      index+=1
  agenda[index] = [nome, telefone, email, twitter, instagram]
```

```python
#Criando função para adicionar dois ou mais contatos de uma vez

def adicionarMultiplosContatos():
  num_cont = int(input('Digite a quantidade de contatos a serem inseridos: '))
  for num in range(num_cont):
    nome = input('\n' + 'Digite o nome do contato: ')
    telefone = input('Digite seu número de telefone: ')
    email = input('Digite seu e-mail: ')
    twitter = input('Digite seu Twitter: ')
    instagram = input('Digite seu Instagram: ')
    index = 1
    for n in agenda.keys():
      if n == index:
        index+=1
    agenda[index] = [nome, telefone, email, twitter, instagram]
```

```python
#Menu da agenda
while True:
  print('\n'+'-'*10 ,'Menu de opções', '-'*10,'\n')
  print('1-Inserir um novo contato')
  print('2-Inserir mais de um contato: ')
  print('3-Consultar um contato cadastrado')
  print('4-Remover um contato cadastrado')
  print('5-Alterar um contato cadastrado')
  print('6-Sair')
  print('\n' + '-'*36)

  controle = int(input('\n' + 'Escolha uma opção: '))
  print('\n' + '-'*36)

  if (controle == 1):
    adicionarContato()
  elif (controle == 2):
    adicionarMultiplosContatos()
  elif (controle == 3):
    consultarContato()
  elif (controle == 4):
    apagarContato()
  elif (controle == 5):
    alterarContato()
  elif (controle == 6):
    break
    
print('\n' + '-'*12,'Até mais!','-'*13)
```

```python
#Criando função para salvar dados dos usuarios cadastrados

def salvarContatos():

  import csv

  dados = 0

  for i in agenda.keys():
    dados = ("%s,%s,%s, %s,%s" % (agenda[i][0], agenda[i][1], agenda[i][2], agenda[i][3], agenda[i][4]))

  arq = open('cadastros.csv', 'w')

  with open('cadastros.csv', 'w') as arq:
    writer = csv.writer(arq)
    writer.writerows(dados)
```
